```python
df_temp["Date"] = pd.to_datetime(df_temp["Date"])

df_temp["Year"] = df_temp["Date"].dt.year
df_temp["DayOfYear"] = df_temp["Date"].dt.strftime("%m-%d")
df_temp = df_temp[df_temp["DayOfYear"] != "02-29"]

temp_2005_2014 = df_temp[(df_temp["Year"] >= 2005) & (df_temp["Year"] <= 2014)]
temp_2015 = df_temp[df_temp["Year"] == 2015]

record_highs = temp_2005_2014[temp_2005_2014["Element"] == "TMAX"].groupby("DayOfYear")["Data_Value"].max()
record_lows = temp_2005_2014[temp_2005_2014["Element"] == "TMIN"].groupby("DayOfYear")["Data_Value"].min()

highs_2015 = temp_2015[temp_2015["Element"] == "TMAX"].groupby("DayOfYear")["Data_Value"].max()
lows_2015 = temp_2015[temp_2015["Element"] == "TMIN"].groupby("DayOfYear")["Data_Value"].min()


broken_highs = highs_2015[highs_2015 > record_highs]
broken_lows = lows_2015[lows_2015 < record_lows]


plt.figure(figsize=(12, 6))
days = np.arange(len(record_highs))

plt.plot(days, record_highs, label="Record High (2005-2014)", color="red", linewidth=1)
plt.plot(days, record_lows, label="Record Low (2005-2014)", color="blue", linewidth=1)
plt.fill_between(days, record_highs, record_lows, color="grey", alpha=0.2)


plt.scatter(broken_highs.index.map(lambda x: list(record_highs.index).index(x)), broken_highs, color="darkred", label="Broken Highs (2015)", zorder=3)
plt.scatter(broken_lows.index.map(lambda x: list(record_lows.index).index(x)), broken_lows, color="darkblue", label="Broken Lows (2015)", zorder=3)


plt.xlabel("Day of the Year")
plt.ylabel("Temperature (Tenths of °C)")
plt.title("Record High and Low Temperatures (2005-2014) with 2015 Breaks")
plt.legend()
plt.show() 
```

```python
import geopandas as gpd
import contextily as ctx


stations_df = df_bin[["ID", "LATITUDE", "LONGITUDE", "NAME"]].dropna()


gdf = gpd.GeoDataFrame(stations_df, geometry=gpd.points_from_xy(stations_df.LONGITUDE, stations_df.LATITUDE), crs="EPSG:4326")


fig, ax = plt.subplots(figsize=(10, 6))
gdf = gdf.to_crs(epsg=3857)  # Convert to Web Mercator projection

gdf.plot(ax=ax, markersize=50, color="red", alpha=0.6, label="Weather Stations", edgecolor="black")
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

ax.set_title("Weather Stations Near Ann Arbor, Michigan")
ax.legend()
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
